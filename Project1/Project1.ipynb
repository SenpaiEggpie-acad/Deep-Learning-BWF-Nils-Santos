```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
census_df = pd.DataFrame()
temp_list = []
for sheet in glob.glob('US_Census Data Cleaning/*.csv'):
    print(sheet)
    temp = pd.read_csv(sheet, usecols=[*range(1,11)])
    temp_list.append(temp)

    
```

```python
census_df = pd.concat(temp_list)
```

```python
#census_df=census_df.reset_index().drop(columns='index')
census_df=census_df.drop_duplicates()
census_df=census_df.reset_index().drop(columns='index')
census_df
```

```python
census_df.describe(include="all")
```

```python
temp_male = []
temp_female = []
combi_entry = []
census_df['GenderPop']
for entry in census_df['GenderPop']:
    combi_entry = entry.split("_")
    temp_male.append(int(combi_entry[0].rstrip(combi_entry[0][-1])))
    #print(combi_entry)
    
```

```python
temp_female = census_df['TotalPop'] - temp_male

census_df = census_df.drop(columns='GenderPop')
census_df['MalePop'], census_df['FemalePop'] = temp_male, temp_female
census_df['MalePop'], census_df['FemalePop'] = census_df['MalePop'].astype(float), census_df['FemalePop'].astype(float)
census_df
```

```python
census_df['Income'] = census_df['Income'].astype(str)

temp_income = []
for entry in census_df['Income']:
    entry = entry.strip("$")
    temp_income.append(float(entry))

census_df['Income'] = temp_income

census_df
```

```python
fig = plt.scatter(census_df['FemalePop']/1000000,census_df['Income']/1000)
plt.title('Female Population (in millions) vs. Income (in thousands)')
plt.xlabel('Female Pop.')
plt.ylabel('Income')
plt.show()
```

```python
races = ['Hispanic','White','Black','Native','Asian','Pacific']

for race in races:
    census_df[race] = census_df[race].astype(str)
    temp_race = []
    for entry in census_df[race]:
        entry = float(entry.strip("%"))
        temp_race.append(entry)
    census_df[race] = temp_race

census_df
```

```python
census_df['Pacific']=census_df['Pacific'].fillna(census_df['Pacific'].mean())
#census_df['Pacific'].mean()
```

```python
for race in races:
    fig = plt.hist(census_df[race])
    plt.title('Population Histogram for {race} in the United States'.format(race=race))
    plt.xlabel('Percentage')
    plt.ylabel('Frequency')
    plt.show()
```
